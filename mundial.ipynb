## funcion de prediccion

### Funcion de poisso

Se escogio la distribucion de poisso ya que es una distribucion discreta que describe el numero de eventos que ocurren en un intervalo de tiempo fijo o region de oportunidad.

Goal: Evento que puede ocurrir en los 90 minutos de un partido de futbol.

### Condiciones de una distribucion de poisso

El numero de eventos se puede contar "no puede haber medio gol", la ocurrencia de eventos son independientes "la ocurrencia de un gol no afecta la probabilidad de otro gol - aunque es debatible - un gol puede ocasionar que otro equipo tenga mas ganas de anotar pero se supone que no afecta", la tasa a la que ocurren los eventos es constante "la probabilidad de que ocurra un gol en 90 minutos es la misma en la que ocurra en otro partido de 90 minutos"y dos eventos no pueden ocurrir exactamente en el mismo instante de tiempo. "dos goles no pueden ocurrir en el mismo instante de tiempo"

P (X=x) = (λ^x * e^λ )/ x!


λequipoA: numero esperado de eventos por intervalo de tiempo


λequipoB: promedio de goles en 90 minutos (Equipo A y Equipo B)


x: el numero de goles en un partido que podria marcar el equipo A o el equipo B

## Funcion predict points

```python
#ESTE EN PAUSA POR AHORA
def predict_points(home, away):
    if home in df_team_strength.index and away in df_team_strength.index: #ambos equipos en la de poderio
        # goals_scored * goals_conceded "esto se ve en las dos filas de abajo" ->multiplico la fortaleza del local por la debilidad del visitante en terminos de goles
        lamb_home = df_team_strength.at[home,'GoalsScored'] * df_team_strength.at[away,'GoalsConceded'] #funcion de poisso "el lambda" para equipo local
        lamb_away = df_team_strength.at[away,'GoalsScored'] * df_team_strength.at[home,'GoalsConceded'] #funcion de poisso "el lambda" para equipo visitante
        # mientras mas goles reciba el equipo visitante "el denominador es mas grande" va a penalizar el equipo visitante y la cantiddad de goles del local va a favorecer al local
        prob_home, prob_away, prob_draw = 0, 0, 0 #se crean tres variables de probabilidades (local, visitante y empate), se inicializan con valor de 0
        # lo que sigue es un bucle for para simular los posibles resultados
        for x in range(0,11): #number of goals home team (1-0,2-0,3-0,...,10-0)
            for y in range(0, 11): #number of goals away team (0-1,0-2,...,0-10) -> (1-1,1-2,1-3...)
                p = poisson.pmf(x, lamb_home) * poisson.pmf(y, lamb_away) #se calcula usando la p de la distribucion de poisso, x=#goles
                ## se acumula esa probabilidad en tres posibles escenarios, ese p anterior se acumula en uno de los escenarios, no en todos
                if x == y:
                    prob_draw += p #empate
                elif x > y:
                    prob_home += p #gana local
                else:
                    prob_away += p #gana visitante
        ### se calcula el puntaje
        points_home = 3 * prob_home + prob_draw #probabilidad 100% o 1 "3 puntos"
        points_away = 3 * prob_away + prob_draw
        return (points_home, points_away) #devuelve el puntaje de equipo local y visitante
    else:
        return (0, 0) #en caso de que falle, devuelve la tupla con valores 0,0, no hay puntaje, queda en nulo, puede que haya equipos que no han participado en un mundial, puede que no haya data historica
```

```python
## ahora se corren todos los partidos de la fase de grupo y se actualiza las tablas de cada grupo
for group in dict_table: #bucle for al diccionario que tiene las tablas, obtengo el nombre de cada grupo
    teams_in_group = dict_table[group]['Team'].values #obtengo el nomnre de los equipos de cada grupo
    df_fixture_group_6 = df_fixture_group_48[df_fixture_group_48['home'].isin(teams_in_group)] #Divido 48 partidos en grupos de 6, es decir, por grupo.
    for index, row in df_fixture_group_6.iterrows(): #ITERACION POR CADA FILA QUE ESTA EN EL DATAFRAME
        home, away = row['home'], row['away'] #Selecciono el equipo local y el visitante y le doy su respectivo nombre
        points_home, points_away = predict_points(home, away) #aplico la formula que he creado previamente, la cual me recibe un equipo local y uno visitante
        # lo anterior me arroja el punto del local y el del visitante
        dict_table[group].loc[dict_table[group]['Team'] == home, 'Pts'] += points_home #agrego los puntos a la tabla de posiciones, equipo local
        dict_table[group].loc[dict_table[group]['Team'] == away, 'Pts'] += points_away #lo mismo para visitantes
    # arreglos a las tablas, orden de acuerdo a los puntoT
    dict_table[group] = dict_table[group].sort_values('Pts', ascending=False).reset_index() #segun col pts
    dict_table[group] = dict_table[group][['Team', 'Pts']]
    dict_table[group] = dict_table[group].round(0) #redondeamos a cero los valores para que no aparezca decimales
```

```python
# ahora se actualiza la tabla:
#Actualizar el fixture de octavos con el 1 puesto (group winner) y 2 puesto (runners up)
for grupo in dict_table:
    group_winner = dict_table[grupo].loc[0, "Team"] #localizar 1 puesto, index 0 y que se encuentre en la col "team"
    runners_up = dict_table[grupo].loc[1, "Team"] #segunda fila en esa tabla
    df_fixture_knockout.replace({f'Winners {grupo}':group_winner,
                                f'Runners-up {grupo}':runners_up}, inplace=True) #se cambia solo la primera fila, por eso se le pone el group
df_fixture_knockout["winner"] = "?" #hasta este momento sabemos quienes se enfrentan pero no quien gana aun
```

```python
# crear funcion get_winner, me permitira conocer el ganador del partido
#parecida a la de predecir de antes
def get_winner(df_fixture_updated): #obtener ganador, del dataframe del fixture actualizado
    for index, row in df_fixture_updated.iterrows(): #iteracion por cada fila del dataframe
        home, away = row['home'], row['away'] #dame el nombre del equipo local y el del visitante
        points_home, points_away = predict_points(home, away) #puntos del equipo local y los puntos del visitante
        if points_home > points_away:
            winner = home
        else:
            winner = away
        df_fixture_updated.loc[index, 'winner'] = winner #localizamos elemento particular que esta en el signo de la interrogacion
    return df_fixture_updated
#ya no devolvemos puntos, solo el nombre de quien gana
```

```python
def update_table(df_fixture_round_1, df_fixture_round_2): #recibe dos inputs, el de la ronda de octavos y el de cuartos
    # uno esta en consecuencia del otro
    for index, row in df_fixture_round_1.iterrows(): #iteracion del fixture de la ronda 1
        winner = df_fixture_round_1.loc[index, 'winner'] #nombre ganador
        match = df_fixture_round_1.loc[index, 'score'] #partido "identificador del partido", para saber de que partido viene
        df_fixture_round_2.replace({f'Winners {match}':winner}, inplace=True) #reemplazar a todo lo que diga winner y luego el numero del partido, y luego reemplazalo por el ganador
    df_fixture_round_2['winner'] = '?' #fixture de cuartos de final
    return df_fixture_round_2
```
